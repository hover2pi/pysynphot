# JWST Filter Support in `pysynphot`

Here's a quick look under the hood for an HST filter:

```python
import pysynphot as ps
from astropy.io import fits
import matplotlib.pyplot as plt
path = '/grp/jwst/jwstetc/pandeia_data/jwst/miri/filters/miri_imager_fnd.fits'
data = fits.getdata(path)
bp = ps.ArrayBandpass(data['WAVELENGTH'], data['THROUGHPUT'])
plt.show(plt.plot(bp.wave, bp.throughput))
```
